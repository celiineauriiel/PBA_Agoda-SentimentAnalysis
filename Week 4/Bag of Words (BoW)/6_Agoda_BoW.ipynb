Pada cell ini dilakukan analisis frekuensi kata untuk mengetahui kata-kata yang paling sering muncul dalam corpus. Pertama, seluruh kata dari setiap kalimat dalam `corpus` dikumpulkan ke dalam satu list bernama `all_words`.

Selanjutnya, digunakan fungsi `Counter` dari library `collections` untuk menghitung jumlah kemunculan setiap kata. Hasil perhitungan ini disimpan dalam variabel `freq`.

Terakhir, fungsi `most_common(20)` digunakan untuk menampilkan 20 kata dengan frekuensi tertinggi. Hasil ini digunakan untuk mengidentifikasi kata-kata dominan dalam dataset, yang kemudian dapat membantu dalam menentukan stopwords tambahan serta memahami pola utama dalam ulasan pengguna.


```python
from collections import Counter

all_words = []

for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

freq = Counter(all_words)

print(freq.most_common(20))
```

# Step 2: List all unique words

Cell ini digunakan untuk menampilkan hasil preprocessing berupa `corpus`, yaitu kumpulan kalimat yang telah dibersihkan dari noise, stopwords, dan telah melalui proses lemmatization.

```python
print("Corpus:", corpus)
```

Kode ini digunakan untuk mengambil semua kata unik (unique words) dari corpus dengan menghilangkan duplikasi, sehingga membentuk vocabulary yang akan digunakan dalam proses representasi teks selanjutnya.


```python
# Finding unique words
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print("Unique words:", unique_words)
```

#Step 3: Create a dictionary with mapping of words to a number. This should now be sorted on frequency of occurrence in descending order.

Finding unique words and their frequencies

Cell ini digunakan untuk menghitung frekuensi kemunculan setiap kata dalam corpus, kemudian mengurutkannya dari yang paling sering muncul, serta membuat pemetaan kata ke angka berdasarkan frekuensinya.


```python
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

word_frequencies = Counter(all_words)

# Print the word counts
print("Word Frequencies:", word_frequencies)

# Sorting words by frequency (descending)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Creating the dictionary mapping words to numbers
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1  # Assign numbers starting from 1

print("Word to Number Mapping (Sorted by Frequency):", word_to_number)
```

Cell ini digunakan untuk mengubah teks dalam corpus menjadi representasi numerik menggunakan metode Bag of Words dengan `CountVectorizer`. Parameter `np.set_printoptions` seperti `threshold`, `edgeitems`, dan `linewidth` digunakan untuk memperbesar tampilan array agar tidak terpotong (tidak muncul `...`), sehingga hasil matriks dapat terlihat lebih jelas saat ditampilkan.


```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

np.set_printoptions(
    threshold=10000,
    edgeitems=20,
    linewidth=200
)

cv = CountVectorizer()
independentFeatures = cv.fit_transform(corpus).toarray()

independentFeatures
```

Cell ini digunakan untuk melihat dimensi matriks hasil Bag of Words, di mana (124, 481) menunjukkan bahwa terdapat 124 kalimat dalam corpus dan 481 kata unik sebagai fitur.


```python
print(independentFeatures.shape)
```

Cell ini digunakan untuk menampilkan kata-kata yang muncul (non-zero values) pada beberapa kalimat dalam corpus beserta jumlah kemunculannya, sehingga memudahkan interpretasi hasil Bag of Words.


```python
feature_names = cv.get_feature_names_out()

for i in range(3):
    print("Sentence:", corpus[i])
    print("Non-zero values:")

    for j, val in enumerate(independentFeatures[i]):
        if val > 0:
            print(feature_names[j], ":", val)

    print("-----")
```
